# 도커를 사용하는 이유
- 서버, 패키지 버전, 운영체제 등등에 따라 프로그램 설치 과정 중 많은 에러가 발생하게 됨
- 또한 설치 과정이 다소 복잡함
- 이러한 문제를 해결하기 위해 나타난 것이 도커

```
1. 도커는 어떤 프로그램을 다운 받는 과정을 매우 간단하게 만들어 준다.
2. 도커는 컨테이너를 사용하여 응용 프로그램을 더 쉽게 제작, 배포, 실행할 수 있도록 설계된 도구이다.
3. 도커는 컨테이너 기반의 오픈소스 가상화 플랫포이며 생태계이다.
```

# 컨테이너란?

### 서버에서의 컨테이너 개념
- redis, 웹팩, mysql 등 다양한 프로그램과 실행 환경을 컨테이너로 추상화
- 동일한 인터페이스 제공
- 프로그램 배포 및 관리 단순화

```
일반 컨테이너 개념에서 물건을 손쉽게 운송해주듯, 프로그램을 손쉽게 이동/배포/관리할 수 있음
AWS, Azure, Google cloud 등 어디에서든 실행 가능
```

### 도커 이미지와 도커 컨테이너 정의

---

##### 📌 컨테이너 = 간단하고 편리하게 프로그램을 실행시켜주는 것
- 컨테이너는 코드와 모든 종속성을 패키지화하여 응용 프로그램이 ```한 컴퓨팅 환경에서 다른 컴퓨팅 환경으로 빠르고 안정적으로 실행되도록 하는``` 소프트웨어의 표준 단위

---

##### 📌 컨테이너 이미지
- 컨테이너 이미지는 코드, 런타임, 시스템 도구, 시스템 라이브러리 및 설정과 같은 ```응용 프로그램을 실행하는 데 필요한 모든 것을 포함하는``` 가볍고 독립적이며 실행 가능한 소프트웨어 패키지
- 이미지는 도커 허브에 들어있음
---
```
도커 이미지는 프로그램을 실행하는 데 필요한 설정이나 종속성을 갖고 있다.

우리는 도커 이미지를 이요해 컨테이너를 생성하고,
도커 컨테이너를 이용하여 프로그램을 실행한다.
```
